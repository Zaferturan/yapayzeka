```python
# kullanıcıdan vize, final ve derse devam punlarını ve ağırlıklarını alan ve 
#geçme puanını hesağlayan 90 ve üstü AA
# 80 ve üstü BA
# 70 ve üstü BB
# 60 ve üstü CB
# 50 FF
```

```python
puan_list=[]
agirlik_list=[]
gecme_puani=0
for i in range(0,3):
    print(i+1 ,". puanı giriniz")
    puan=input()
    puan_list.append(float(puan))
    print(i+1 ,". ağıtrlık katsayısını giriniz")
    agirlik=input()
    agirlik_list.append(float(agirlik))
for i in range(0,3):
    gecme_puani=gecme_puani+puan_list[i]*agirlik_list[i]
print(gecme_puani)
```
