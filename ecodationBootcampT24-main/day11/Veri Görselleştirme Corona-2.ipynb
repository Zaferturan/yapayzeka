```python
df['date']=pd.to_datetime(df['date'])
df['date'].dtype
```

```python
df_sec=df[['date', 'Turkey', 'China', 'Italy','Germany','US']]
df_sec
```

```python
df_sec.set_index('date',inplace=True)
df_sec
```

```python
df_sec.plot();
```

```python
#Kümülatif Toplam
sum_daily=df_sec.cumsum(axis=0) # Sütundaki veriyi topluyor
sum_daily
```

```python
sum_daily.plot()
```

```python
sum_daily['Turkey'].plot()
```

```python
sum_daily['Total']=sum_daily.sum(axis=1)
sum_daily
```
