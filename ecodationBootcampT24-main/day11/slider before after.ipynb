```python
import leafmap.foliumap as lm
img1=r'orman/sonra.jpg'
img2=r'orman/once.jpg'
lm.image_comparison(img1,
                    img2,
                    label1='Sonra',
                    label2='Önce',
                    starting_position=50, #50 olursa tam ortada başlar
                    out_html='orman_once_sonra.html' # bunu bir html dosyası olarak kaydediyoruz
)
```
