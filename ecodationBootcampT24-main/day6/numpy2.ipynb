```python
np.random.seed(42) #random state 42 rastglee diziler içinde belirli bir diziyi seçiyor
notlar1=np.random.normal(loc=50,scale=5,size=20)
notlar1=np.rint(notlar1).astype(int)
print(notlar1)
notlar2=.7*notlar1+np.random.normal(loc=0, scale=5,size=20)
notlar2=np.rint(notlar2).astype(int)
print(notlar2)
corr=np.corrcoef(notlar1,notlar2)
corr
```

### hava sıcaklıkları ile ısıtıcı satışları negatif yönlü korelasyona örnektir
Sigara tüketimi ve sağlık durumu: Sigara tüketiminin artması ile bireyin sağlık durumunun kötüleşmesi arasındaki negatif yönlü korelasyon.

Uyku süresi ve stres düzeyi: Uyku süresinin azalması ile stres düzeyinin artması arasındaki negatif yönlü korelasyon.

Egzersiz süresi ve vücut kitle indeksi: Egzersiz süresinin artması ile vücut kitle indeksinin azalması arasındaki negatif yönlü korelasyon
.
Kar yağışı ve elektrik tüketimi: Kar yağışının artması ile elektrik tüketiminin azalması arasındaki negatif yönlü korelasyo
n.
Faiz oranları ve hisse senedi fiyatları: Faiz oranlarının artması ile hisse senedi fiyatlarının düşmesi arasındaki negatif yönlü korelasy
on.
Gelir düzeyi ve yoksulluk oranı: Gelir düzeyinin artması ile yoksulluk oranının azalması arasındaki negatif yönlü korelasyon.
