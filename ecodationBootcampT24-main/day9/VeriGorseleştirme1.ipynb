```python
df['Date'].dtype # Veri itpi Object
```

```python
df['Date']=pd.to_datetime(df['Date']) # tarih formatına çevir
df['Date'].info()
```

```python
# Tarihten ayın kaçıncı günü olduğunu çıkarma
df['day']=df.Date.dt.day
df['day']
```

```python
df['Date']
```

```python
df['day_name']=df['Date'].dt.day_name()
df['day_name']
```

```python
df['month_name']=df['Date'].dt.month_name()
df['month_name']
```

```python
df['Time'].dtype
```

```python
df['Time']=pd.to_datetime(df['Time'])
df['Time']
```

```python
renkler={'Male':'blue', 'Female':'pink'}
sns.countplot(x=df['Product line'], hue=df['Gender'],
             palette=renkler)
plt.title('Cinsiyete Göre Alış Veriş Türleri')
plt.xticks(rotation=45);
```
